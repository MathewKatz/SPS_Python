```python
messy = pd.read_csv('messy-data.csv')
```

```python
messy
```

```python
messy.columns
```

```python
messy.drop(messy.columns[[0, 4, 5, 6, 7, 8, 9]], axis=1, inplace=True)
```

```python
messy
```

```python
messy=messy.drop(messy.index[14:20])
```

```python
messy
```

```python
messy=messy.drop(messy.index[0:2])
```

```python
messy
```

```python
messy.set_index('Seasonal rainfall in Lake Victoria and Simiyu', inplace = True)
```

```python
messy.head()
```

```python
messy.columns = ['Lake Victoria Seasonal Rainfall in mm', 'Simiyu Seasonal Rainfall in mm']
```

```python
messy
```

```python
messy.index.names = ['Date']
```

```python
messy['Lake Victoria Seasonal Rainfall in mm'] = messy['Lake Victoria Seasonal Rainfall in mm'].map(lambda x: x.rstrip('m'))
```

```python
messy['Simiyu Seasonal Rainfall in mm'] = messy['Simiyu Seasonal Rainfall in mm'].map(lambda x: x.rstrip('m'))
```

```python
messy['Lake Victoria Seasonal Rainfall in mm'] = messy['Lake Victoria Seasonal Rainfall in mm'].astype('float')
```

```python
messy['Simiyu Seasonal Rainfall in mm'] = messy['Simiyu Seasonal Rainfall in mm'].astype('float')
```
